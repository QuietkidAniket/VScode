## How to filter rows from a Pandas DataFrame

```python
movies = pd.read_csv("http://bit.ly/imdbratings")
movies
```

```python
type(False)
```

```python
# filtering the rows using python loop  (method 1)
booleans = []
for length in  movies.duration:
    if length >= 200:
        booleans.append(True)
    else:
        booleans.append(False)
```

```python
booleans[0:5]
```

```python
is_long = pd.Series(booleans)
is_long.head()
```

```python
movies[is_long]   #finally we get the filtered rows
```

```python
# without using for loop (method 2)
is_long = movies.duration >= 200
is_long.head()
```

```python
movies[is_long]
```

### quickest and most shortest and simplest way to filter rows based on a condition : 

```python
movies[movies.duration >= 200]
```

```python
# just the genre : 
movies[movies.duration >=200]['genre']
```

```python
# to allow select rows and columns by labels : 
movies.loc[movies.duration >=200, 'genre']
```

## How do I use the axis parameter?

```python
drinks = pd.read_csv('http://bit.ly/drinksbycountry')
drinks.head()
```

### When referring to rows or columns with the axis parameter:

#### Axis 0 refers to rows
#### Axis 1 refers to columns

```python
# drop a column (temporarily)
drinks.drop('continent', axis=1).head()
```

```python
#  drop a row (temporarily)
drinks.drop(2, axis = 0).head()
```

### When performing a mathematical operation with axis parameter
#### axis 0 means the operation should "move down" the row axis
#### axis 1 means the operation should "move across" the column axis

```python
# calculate the mean of each column
drinks.mean()

#or equivalently specify the axis explicitly
drinks.mean(axis = 0)           # alias for axis = 0 is axis = 'index'
```

```python

# calculate the mean of each row
drinks.mean(axis = 1).head()            # alias for axis = 1 is axis = 'columns'
```

### how to use the groupby in pandas

```python
drinks = pd.read_table("https://bit.ly/drinksbycountry")
drinks.head()
```

```python
# calculate the mean beer servings across the entire dataset
drinks.beer_servings.mean()
```

```python
# calculate the mean beer servings just for countries in Africa
drinks[drinks.beer_servings == "Africa"].beer_servings.mean()
```

```python
# calculate the mean beer servings for each continent
drinks.groupby('continents').beer_servings.mean()                                
```

```python
# multiple aggregate functions may also be added
drinks.groupby('continents').beer_servings.agg(['count', 'mean', 'min', 'max'])
```

```python
# specifying the column to which the aggregation function should be applied is not required
drinks.groupby("continent").mean()
```

```python
# allow plots to appear in the notebook
%matplotlib inline
# side-by-side bar plot of the DataFrame directly above
drinks.groupby("continents").mean().plot(kind= "bar")
```

### Grouping records using groupby

```python
# finding the most common wine reviewers in the data set, counts how many reviews each person wrote
reviews.groupby('taster_twitter_handle').taster_twitter_handle.size()
```

```python
# What is the best wine I can buy for a given amount of money?
# For a certain price, Wines with the max points are shown. Prices increase from top to bottom
reviews.groupby('price').points.max().sort_index()
```

```python
# minimum and maximum prices of each variety of wine
min_max_wine = reviews.groupby('variety').price.agg([min, max])
min_max_wine.head()
```

```python
# let us sort the above dataframe on the basis of the prices
min_max_wine.sort_values(by=['min', 'max'], ascending = False).head()
del min_max_wine
```

```python
# finding the average review score of each individual reviewer
reviews.groupby('taster_name').points.mean()
```

```python
# checking the stats of the average scores assigned by the various reviewers
reviews.groupby('taster_name').points.mean().describe()
```

```python
# what combination of countries and wine variety are most common? Sort them on the basis of their counts
reviews.groupby(['country','variety']).size().sort_values(ascending = False)
```
