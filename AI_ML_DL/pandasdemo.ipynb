## How to filter rows from a Pandas DataFrame

```python
movies = pd.read_csv("http://bit.ly/imdbratings")
movies
```

```python
type(False)
```

```python
# filtering the rows using python loop  (method 1)
booleans = []
for length in  movies.duration:
    if length >= 200:
        booleans.append(True)
    else:
        booleans.append(False)
```

```python
booleans[0:5]
```

```python
is_long = pd.Series(booleans)
is_long.head()
```

```python
movies[is_long]   #finally we get the filtered rows
```

```python
# without using for loop (method 2)
is_long = movies.duration >= 200
is_long.head()
```

```python
movies[is_long]
```

### quickest and most shortest and simplest way to filter rows based on a condition : 

```python
movies[movies.duration >= 200]
```

```python
# just the genre : 
movies[movies.duration >=200]['genre']
```

```python
# to allow select rows and columns by labels : 
movies.loc[movies.duration >=200, 'genre']
```

## How do I use the axis parameter?

```python
drinks = pd.read_csv('http://bit.ly/drinksbycountry')
drinks.head()
```

### When referring to rows or columns with the axis parameter:

#### Axis 0 refers to rows
#### Axis 1 refers to columns

```python
# drop a column (temporarily)
drinks.drop('continent', axis=1).head()
```

```python
#  drop a row (temporarily)
drinks.drop(2, axis = 0).head()
```

### When performing a mathematical operation with axis parameter
#### axis 0 means the operation should "move down" the row axis
#### axis 1 means the operation should "move across" the column axis

```python
# calculate the mean of each column
drinks.mean()

#or equivalently specify the axis explicitly
drinks.mean(axis = 0)           # alias for axis = 0 is axis = 'index'
```

```python

# calculate the mean of each row
drinks.mean(axis = 1).head()            # alias for axis = 1 is axis = 'columns'
```
