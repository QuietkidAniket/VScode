```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.plot([1,2,3,4])
plt.show()
```

```python
plt.figure(figsize= (5,5))
plt.plot([1,2,3,4], [1,4,9,16], "rx")       # x,y, "red x symbol"
plt.title("Test plot", fontsize=8, color ='g')
plt.axis([0,6,0,20 ])   # x will be plotted from 0 to 6 and y will be plotted from 0 to 20
plt.xlabel("Number M")
plt.annotate("square it", (3,6))
plt.ylabel("n^2")
```

### Bar chart

```python
plt.clf()   # clear the figure
x = np.arange(4)
y = [8.8, 5.2, 3.6, 5.9]
plt.xticks(x , ("Ankit", "Hans", "Joe", "Flanco"))
plt.bar(x, y, color = [ 'lime', "r", "k", "tan"])
plt.show()
```

```python
d = {
    "Red O" : np.random.rand(10),
    "Grn X" : np.random.rand(10)   
}
df = pd.DataFrame(d)
df.plot(style=['ro', 'gx'])         # DataFrames are actually able to call the plot function of matplotlib
plt.show()
```

### Random dots in a scatter

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
sizes = (30 * np.random.rand(N))**2    # 0 to 15 radii
plt.scatter(x,y, s= sizes, c =  colors,alpha =0.5) 
plt.show() 
```

```python
fig = plt.figure()
fig.suptitle("My subplots")
fig.add_subplot(221)
plt.plot([np.log(n) for n in range(1,10)])   
fig.add_subplot(222, facecolor= 'y')
fig.add_subplot(223)
fig.add_subplot(223)
plt.show()
```

```python
fig, plots = plt.subplots(2, sharex= True)
fig.suptitle("Sharing X axis")
x = range(0,200,5)
y = [n**0.8 for n in x]
plots[0].plot(x,y, color= 'r')
plots[1].scatter(x,y)
plt.show()  
```
