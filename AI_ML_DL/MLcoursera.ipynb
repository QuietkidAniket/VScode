```python
def compute_model_output(x, w, b):
  """
    Computes the prediction of a linear model
    Args:
      x (ndarray (m,)): Data, m examples
      The argument description (ndarray (m,)) describes a Numpy n-dimensional array of shape (m,).

      w,b (scalar)    : model parameters
      (scalar) describes an argument without dimensions, just a magnitude.
    Returns
      y (ndarray (m,)): target values
  """
  m = x.shape[0]
  # np.zeros(n) will return an one dimensional numpy array with n entries
  f_wb = np.zeros(m)
  for i in range(m):
    f_wb[i] = w * x[i]  + b
  return f_wb
```

```python
#predicted weight and bias values (intuition)
w = 200
b = 100
tmp_f_wb = compute_model_output(x_train, w, b,)

#plot our model prediction
plt.plot(x_train, tmp_f_wb, c='b', label='Our prediction')

#plot the data points
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual values')

#set the title
plt.title("Housing Prices")
#set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
#set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```

```python
#gradient descent algorithm implementation
def compute_gradient(x, y, w, b):
  """
  Computes the gradient for the linear regression
  Args:
      x (ndarray (m,)): Data, m examples / features
      y (ndarray (m,)): target values
      w,b (scalar)    : model parameters
    Returns
      dj_dw (scalar): The gradient of the cost w.r.t. the parameters w
      dj_db (scalar): The gradient of the cost w.r.t. the parameters b
  """
  #number of training examples
  m = x.shape[0]
  dj_dw = 0
  dj_db = 0
  for i in range(m):
    f_wb = w * x[i] + b
    dj_w_i = (f_wb - y[i]) * x[i]
    dj_b_i = f_wb - y[i]
    dj_dw += dj_w_i
    dj_db += dj_b_i
  dj_dw = dj_dw/m
  dj_db = dj_db/m

  return dj_dw, dj_db

```

### Compute cost using multiple variables

```python
def compute_cost(x,y,w,b):
        """
        computes cost
        Args:
        x   (ndarray (m,n)): Data, m examples with n features
        y   (ndarray (m, )): target values
        w   (ndarray (n, )): model parameters
        b   (scalar )      : model parameter
        """
        m = x.shape[0]     
        cost = 0.0
        for i in range(m):
            f_wb_i = np.dot(x[i], w) +b
            cost += (f_wb_i - y[i])**2
        cost /= 2*m
        return cost
```

### Gradient descent with multiple variables

```python
def compute_gradient( x,y,w,b): 
    """
    computes the gradient for all linear regression
    Args:
    X (ndarray (m,n)): Data, m examples with n features
    Y (ndarray (m,)) : target values
    w (ndarray (n,)) : model parameters  
    b (scalar)       : model parameter
    Returns:
    dj_dw (ndarray(n,)) : Gradient of the cost w.r.t. the parameters of w
    dj_db (scalar)      : Gradient of the cost w.r.t. the parameters of b
    """
    m,n = x.shape                   # number of examples, no. of features
    dj_dw = np.zeros((n,))
    dj_db = 0.
    for i in range(m):
        err = (np.dot(x[i], w) + b) -  y[i]
        for j in range(n):
            dj_dw +=  err * x[i,j]
        dj_db += err
    dj_dw /= m
    dj_dw /= m
    return dj_db, dj_dw
```

```python
def gradient_descent(X, y, w_in, b_in, cost_function, gradient_function, alpha, num_iters): 
    """
    Performs batch gradient descent to learn w and b. Updates w and b by taking 
    num_iters gradient steps with learning rate alpha
    
    Args:
      X (ndarray (m,n))   : Data, m examples with n features
      y (ndarray (m,))    : target values
      w_in (ndarray (n,)) : initial model parameters  
      b_in (scalar)       : initial model parameter
      cost_function       : function to compute cost
      gradient_function   : function to compute the gradient
      alpha (float)       : Learning rate
      num_iters (int)     : number of iterations to run gradient descent
      
    Returns:
      w (ndarray (n,)) : Updated values of parameters 
      b (scalar)       : Updated value of parameter 
      """
    
    import copy, math
    # An array to store cost J and w's at each iteration primarily for graphing later
    J_history = []
    w = copy.deepcopy(w_in)  #avoid modifying global w within function
    b = b_in
    
    for i in range(num_iters):

        # Calculate the gradient and update the parameters
        dj_db,dj_dw = gradient_function(X, y, w, b)   ##None

        # Update Parameters using w, b, alpha and gradient
        w = w - alpha * dj_dw               ##None
        b = b - alpha * dj_db               ##None
      
        # Save cost J at each iteration
        if i<100000:      # prevent resource exhaustion 
            J_history.append( cost_function(X, y, w, b))

        # Print cost every at intervals 10 times or as many iterations if < 10
        if i% math.ceil(num_iters / 10) == 0:
            print(f"Iteration {i:4d}: Cost {J_history[-1]:8.2f}   ")
        
    return w, b, J_history #return final w,b and J history for graphing
```
