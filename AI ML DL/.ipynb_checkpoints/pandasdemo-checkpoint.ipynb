# Introduction to Pandas 

```python
import pandas as pd
```

```python
orders = pd.read_table("http://bit.ly/chiporders")
```

```python
orders.head() # returns a specified number of rows from the top
```

```python
movie_users = pd.read_table("http://bit.ly/movieusers")
```

```python
movie_users.head()   
# two problems encountered here : 
# 1> the delimiter used in the file is | (pipeline), but the default delimiter for read_table is tab
# 2> pandas read the first row of data as the header of the table which in this case is just a data row
```

```python
# Solving the above mentioned problems : 
movie_users = pd.read_table("http://bit.ly/movieusers", sep = '|', header = None)
movie_users.head()
```

```python
# setting the table header
user_cols = ["user_id", "age", "gender", "occupation", "zip_code"]
movie_users = pd.read_table("http://bit.ly/movieusers", sep = '|', header = None, names = user_cols)
movie_users.head()
```

## Introduction to DataFrame

```python
ufo = pd.read_table("http://bit.ly/uforeports", sep = ',') 
# an alternative for the above statement is  ufo = pd.read_csv("http://bit.ly/")
```

```python
# confirming that 'ufo' is a DataFrame
type(ufo)
```

```python
# to see the first five rows : 
ufo.head()
```

```python
# Selecting a Panda Series from a DataFrame
print(ufo['City'], '\n')  # Selecting the city Series from ufo DataFrame
print(type(ufo['City']))  # Confirming whether its a Series
```

```python
# Everytime a Series is added to a DataFrame, its name becomes an attribute of the DataFrame
ufo.City
# BUT AVOID USING ATTRIBUTES
```

```python
# Analogous to concatenation of strings ( 'ab'  + 'cd' = 'abcd')
# We can 'concatenate' Series in the way we prefer
ufo["Location"] = ufo["City"] + ', ' + ufo["State"]
ufo.head()
```

## How do I rename columns in Pandas DataFrame

```python
ufo = pd.read_csv("http://bit.ly/uforeports")
```

```python
ufo.head() # let us look at the frist 5 rows
```

```python
# Viewing all the columns : 
ufo.columns
```

```python
# renaming columns using 1st method : 
ufo.rename(columns = {"Colors Reported": "Colors_Reported"}, inplace = True)
```

```python
ufo.head()
```

```python
# Renaming Columns using 2nd method :
ufo_cols = ["city", "colors_reported", "shape_reported", "state", "time"]
ufo.columns = ufo_cols
ufo.head()
```

```python
# Renaming the columns using 3rd method : 
ufo = pd.read_csv("http://bit.ly/uforeports", header = 0, names = ufo_cols) 
ufo.head()
```

## How to remove a column from Pandas DataFrame

```python
ufo = pd.read_csv("http://bit.ly/uforeports")
ufo.head()
```

```python
# deleting a single column
ufo.drop('Colors Reported', axis= 1, inplace =True)
```

```python
ufo.head()
```

```python
ufo.drop(["City", "State"], axis = 1, inplace = True)
ufo.head()
```

```python
# dropping two rows (the two entries of 1930 )from the first :
ufo.drop([0,1], axis = 0, inplace = True)
ufo.head()
```

## How do I sort pandas DataFrame or Series

```python
movies = pd.read_csv("http://bit.ly/imdbratings")
movies.head()
```

```python
movies['title'].sort_values(ascending = False)  # sorts the elements in descending order of their title name
```

```python
# but sorting do not permanently affect the underlying data
movies['title']
```

```python
# sorting 2nd method"
movies.sort_values('title')
```

```python
movies.head()
```

```python
# sorting rows based on two columns : 
movies.sort_values(['genre', 'star_rating'])
```
